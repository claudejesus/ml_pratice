```python
import gradio as gr
```

```python
import pickle
```

```python
my_model = pickle.load(open('mymodel.pkl', 'rb'))
my_model
```

```python
def make_prediction(Age, Salary, Education, Capital):
    
    pred = my_model.predict([[Age, Salary, Education, Capital]])
    if pred > 0.5:
        output = "You are eligible for the loan! Congratulation"
    else:
        output = "Sorry, You are not eligible for the loan! "
        
    return (output)


Age = gr.Number(label="Enter your Age")
Salary = gr.Slider(0, 200000, label="Enter your Salary")
Education = gr.Radio([1.,0.], label='Selecct your Education level')
Capital = gr.Number(label="Enter the Capital ")

output = gr.TextArea()

interface = gr.Interface(fn=make_prediction, inputs=[Age, Salary, Education, Capital], outputs=output)

interface.launch(share=True)
        
        
```
